## one-Hot Encoding

####  One-Hot encoding is used for categorical columns. Because both infill pattern and material categorical features are not ordinal categorical columns. Label encoding is used can for ordinal categorical columns.

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply Label Encoding to each categorical column
DataFrame_encoded = DataFrame.copy() # Create a copy to avoid modifying the original DataFrame
for col in categorical_cols:
    DataFrame_encoded[col] = label_encoder.fit_transform(DataFrame_encoded[col])

print("DataFrame after Label Encoding:")
display(DataFrame_encoded.head())
```

### Data splitting

```python

x = DataFrame_encoded.drop(['material'], axis=1)
y = DataFrame_encoded['material']
```

##### Data is splitted from the DataFrame 'DataFrame_encoded' for two variables x(independent variable) and y(dependent variable).
##### Dependent variable is nothing but output in the dataset and independent variable is remaining all features

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.shape
```

### Model Building

```python
models = {
    "RandomForest": {
        "model": RandomForestClassifier(random_state=42),
        "params": {
            "n_estimators": [100, 200, 300, 400, 500],
            "max_depth": [4, 6, 8, 10, 12, None],
            "min_samples_split": [2, 4, 6, 8, 10],
            "min_samples_leaf": [1, 2, 3, 4, 5, 7],
            "max_features": ["sqrt", "log2", None],
        }
    },
    "XGBoost": {
        "model": XGBClassifier(random_state=42, eval_metric='logloss', use_label_encoder=False),
        "params": {
            "n_estimators": [100, 200, 300, 400, 500],
            "learning_rate": [0.01, 0.05, 0.1, 0.2, 0.3],
            "max_depth": [3, 5, 7, 9, 10, 12],
            "subsample": [0.6, 0.7, 0.8, 0.9, 1.0],
            "colsample_bytree": [0.6, 0.7, 0.8, 0.9, 1.0],
            "reg_lambda": [1, 1.5, 2]
        }
    },
    "LogisticRegression": {
        "model": LogisticRegression(random_state=42, max_iter=1000),
        "params": {
            "C": [0.01, 0.1, 0.5, 0.7, 1.0, 2.0, 5.0],
            "solver": ["lbfgs", "liblinear", "saga"],
            "penalty": ["l1", "l2", "elasticnet", "none"],
            "l1_ratio": [0, 0.25, 0.5, 0.75, 1]  # only used if penalty="elasticnet"
        }
    }
}
```

```python
best_models = {}

for name, mp in models.items():
    print(f"\n===== {name} Grid Search =====")
    grid = GridSearchCV(mp["model"], mp["params"], cv=5, scoring='accuracy', n_jobs=-1)
    grid.fit(x_train, y_train)
    
    best_models[name] = grid.best_estimator_
    print(f"Best Parameters: {grid.best_params_}")

    y_train_pred = grid.predict(x_train)
    y_test_pred = grid.predict(x_test)
    
    # Metrics
    print(f"============={name}===========")
    print(f"Train Accuracy: {accuracy_score(y_train, y_train_pred):.4f}")
    print(f"Test Accuracy : {accuracy_score(y_test, y_test_pred):.4f}")
    print("\nConfusion Matrix (Test):")
    print(confusion_matrix(y_test, y_test_pred))
    print("\nClassification Report (Test):")
    print(classification_report(y_test, y_test_pred))
```

```python
print("\n===== Summary (Test Accuracy) =====")
for name, model in best_models.items():
    y_pred = model.predict(x_test)
    print(f"{name}: {accuracy_score(y_test, y_pred):.4f}")
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name, mp in models.items():
    print(f"\n===== {name} Stratified CV =====")
    estimator = mp["model"]
    cv_scores = cross_val_score(estimator, x_train, y_train, cv=skf, scoring='accuracy', n_jobs=-1)
    estimator.fit(x_train, y_train)
    print(cv_scores)
    print(f"CV Accuracy (mean Â± std): {np.mean(cv_scores):.4f} Â± {np.std(cv_scores):.4f}")
```
